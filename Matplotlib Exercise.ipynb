```python
import matplotlib.pyplot as plt
```

```python
#1. Creating a Simple Line Plot
#Question: Create a simple line plot using Matplotlib with the following data:
#x = [1, 2, 3, 4, 5], y = [10, 20, 25, 30, 40]
x= [1,2,3,4,5]
y= [10,20,25,30,40]

plt.plot(y,x)
plt.show()
```

```python
#2. Adding Titles and Labels
#Question: Modify the plot from question 1 to include the title "Simple Line Plot" and labels for the x-axis and y-axis: "X-Axis" and "Y-Axis".
plt.plot(y,x)
plt.title('A Plot of X vs Y')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#3. Customizing Line Styles
#Question: Change the color of the line in the plot from question 1 to 'green' and make the line dashed.
x= [1,2,3,4,5]
y= [10,20,25,30,40]

plt.plot(y,x, color= 'green', linestyle= '--')

plt.title('A Plot of X vs Y')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Real-World Scenario
#Question: You have monthly sales data for a small business:
#months = ['Jan', 'Feb', 'Mar', 'Apr', 'May'], sales = [200, 300, 250, 450, 500].
#Create a line plot to visualize the sales trend over the months.

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [200, 300, 250, 450, 500]

plt.plot(months,sales, color= 'red', linestyle= '--')
plt.title('Sales Trend Over Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
#Day 2: Advanced Matplotlib
#1. Creating a Bar Chart
#Question: Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'], values = [50, 30, 40, 60].

categories= ['A','B','C','D']
values= [50,30,40,60]
plt.bar(categories,values)
plt.title('Category vs Value')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
#2. Creating a Scatter Plot
#Question: Plot a scatter plot for the following data: x = [5, 10, 15, 20, 25], y = [10, 15, 20, 25, 30].

x = [5, 10, 15, 20, 25]
y = [10, 15, 20, 25, 30]
plt.scatter(x,y)
plt.title('Plot X vs Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Subplots
#Question: Create two subplots: a line plot in the first subplot and a bar chart in the second subplot. Use any data of your choice.

fig, axs= plt.subplots(1,2, figsize= (8,4))
#A line plot
x= [1,2,3,4,5]
y= [10,20,25,30,40]
axs[0].plot(x,y, color= 'red')
axs[0].set_title('Line Plot')

#Barchart
z= [200, 300, 250, 450, 500]
axs[1].bar(x,z, color= 'brown')
axs[1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
#4. Real-World Scenario
#Question: A company wants to compare its quarterly profits for two years:
#quarters = ['Q1', 'Q2', 'Q3', 'Q4']
#profits_2022 = [10000, 12000, 11000, 13000]
#profits_2023 = [15000, 14000, 16000, 17000]
#Create a grouped bar chart to visualize the comparison between 2022 and 2023 profits.

#2022 quarterly profit
fig,axs= plt.subplots(1,2, figsize= (10,5))
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
profits_2022 = [10000, 12000, 11000, 13000]
axs[0].bar(quarters,profits_2022, color= 'orange')
axs[0].set_title('2022 Quarterly Profit')

#2023 quarterly profit
profits_2023 = [15000, 14000, 16000, 17000]
axs[1].bar(quarters,profits_2023, color= 'black')
axs[1].set_title('2023 Quarterly Profit')

plt.tight_layout()
plt.show()
```
