# Seaborn

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data= sb.load_dataset('tips')
sb.scatterplot(x= 'total_bill', y= 'tip', data= data)
plt.title('Total Bill vs Tip Amount')
plt.show()
```

```python
#barplot
sb.barplot(x= 'day', y= 'total_bill', data=data)
plt.title('Total Bill by Day of Week')
plt.show()
```

```python
#1. Creating a Simple Histogram
#Question: Use the tips dataset and create a histogram to visualize the distribution of total bill amounts.
sb.histplot(x= 'tip', y= 'total_bill', data=data, color= 'cyan')
plt.title('The Distribution of Total Bill Amount vs Tips Amount')
plt.show()
```

```python
#. Creating a Box Plot
#Question: Use the tips dataset to create a box plot showing the distribution of total bill amounts across different days.
sb.boxplot(x= 'day', y= 'total_bill', data=data, color= 'grey')
plt.title('Distribution of Total Bill Amounts Across Different Days')
plt.show()
```
