# Intoduction to Matplotlib

```python
# import the library
import matplotlib.pyplot as plt
```

```python
# Days vs Stock Prices
days = [1,2,3,4,5]
prices = [700,1200,1500,1400,1800]
```

```python
plt.plot(days,prices)
plt.title('Naira Prices over Days')
plt.xlabel('Days')
plt.ylabel('Prices')
plt.show()
```

```python
#Example 2
categories = ['electronics','furniture','clothing','food']
sales = [20000,15000,5000,10000]
```

```python
plt.bar(categories,sales)
plt.title('Sales by Product')
plt.xlabel('Category')
plt.ylabel('Sales($)')
plt.show()
```

```python
# Subplots as an example
#lets assign two variables
fig,axs = plt.subplots(1,2, figsize=(10,5))

#page view
days= [1,2,3,4,5]
views= [100,120,130,115,150]
axs[0].plot(days,views)
axs[0].set_title('Page View Over 5 Days')

#sign-ups
signups= [5,10,15,7,9]
axs[1].bar(days,signups)
axs[1].set_title('User Sign-ups Over 5 Days')

plt.tight_layout()
plt.show()
```
